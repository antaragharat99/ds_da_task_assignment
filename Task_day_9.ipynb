```python
def greet():
    print("Hello! Welcome to Python.")

greet()  # Calling the function
```

```python
def greet(name):
    print("Hello", name)

greet("Dhiraj")
```

```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
def greet(name="Guest"):
    print("Hello", name)

greet()  # Uses default value
greet("Dhiraj")  # Uses provided value
```

```python
def multiply(a, b):
    print("Result:", a * b)

multiply(4, 5)
```

```python
def say_hello():
    print("Hello!")

say_hello()
```

```python
def add(a, b=5):
    return a + b

print(add(3))
```

```python
add = lambda a, b: a + b
print(add(5, 3))
```

```python
square = lambda x: x * x
print(square(4))
```

```python
maximum = lambda a, b: a if a > b else b
print(maximum(10, 20))
```

```python
is_even = lambda n: "Even" if n % 2 == 0 else "Odd"
print(is_even(7))
```

```python
concat = lambda s1, s2: s1 + " " + s2
print(concat("Hello", "Python"))
```
