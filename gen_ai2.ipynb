```python
### Transformer  - Pipeline as a Module - Ready Made AI Workflow
# Text == Pretrained Model == Useful output
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")  #### This is the function of the Model
data = ["I love you", "I hate you","I love learning Transformers","I am trust you"]
result  = sentiment_pipeline(data)
print(result)
```

```python
classifier = pipeline("text-classification")
classifier("Win a free Iphone now !!")
```

```python
classifier = pipeline("zero-shot-classification")
text = "The stock market crashed due to inflation concerns."
labels = ["finance", "politics", "sports"]
classifier(text, labels)
```

```python
ticket = "My internet connection is very slow."
labels = ["billing issue",'technical support','account management']
classifier(ticket,labels)
```

```python
import pandas as pd
df = pd.DataFrame({
    'ticketid': [1,2,3,4],
    'ticket_reason': [
        'My Internet connectio is very slow.',
         'I was charged twice on my bill for this month.',
        'I forgot my passwrd and cannot login.',
        'How can I upgrade my plan?.'
    ]
})
print(df)
```

```python
labels= ['technical support','billing issue','account management','sales inquiry']
classifier = pipeline("zero-shot-classification")
```

```python
for _, row in df.iterrows():
    result = classifier(row['ticket_reason'],labels)
    print(result)
```
