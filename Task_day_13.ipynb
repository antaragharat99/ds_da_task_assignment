```python
import pandas as pd
data = {
    "Name": ["Amit", "Ravi", "Priya"],
    "Age": [25, 30, 22],
    "City": ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
display(df)
```

```python

print(df.head(1))       # First 5 rows
print(df.tail(2))      # Last 2 rows
print(df.shape)        # (rows, columns)
print(df.columns)      # Column names
print(df.info())       # Summary
print(df.describe())   # Stats summary
```

```python
print(df["Name"])        # Select column
print(df[["Name","Age"]])# Multiple columns
print(df.iloc[0])        # Row by index
print(df.loc[1])         # Row by label
```

```python
print(df[df["Age"] > 23])    # Filter rows
```

```python
import pandas as pd
data = {
"Name": ["Apoorv", "Ravi", "Sneha", "Kiran", "Anjali"],
"marks": [98, 62, 89, 93, 81],
}
df = pd.DataFrame(data)
display(df)
display(df[df["marks"]>80])
display(df["marks"].max())
```

```python
sorted_df = df.sort_values(by="marks", ascending=True)
display(sorted_df)
```

```python
df["Age"] = [15, 16, 14, 17, 15]        # Example ages
df["Age_plus_10"] = df["Age"] + 10

display(df)
```

```python
df["marks_plus_10"] = df["marks"] + 10

display(df)
```

```python
df["City"] = ["Mumbai", "Thane", "Pune", "Navi Mumbai", "Mumbai"]

display(df[df['City']=="Mumbai"])
```
