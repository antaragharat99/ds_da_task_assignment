```python
import numpy as np

data = [10, 20, 30, 40, 50]
mean_value = np.mean(data)
print(mean_value)
```

```python
import numpy as np

data = [10, 20, 30, 40, 50]
median_value = np.median(data)
print(median_value)
```

```python
from statistics import mode

data = [10, 20, 20, 30, 40]
mode_value = mode(data)
print(mode_value)
```

```python
data = [10, 20, 30, 40, 50]
range_value = max(data) - min(data)
print(range_value)
```

```python
import numpy as np

data = [10, 20, 30, 40, 50]
variance_value = np.var(data)
print(variance_value)
```

```python
import numpy as np

data = [10, 20, 30, 40, 50]
std_value = np.std(data)
print(std_value)
```

```python
import pandas as pd

data = {
    "Marks": [60, 70, 80, 90, 100]
}

df = pd.DataFrame(data)

print(df.describe())
```

```python
import numpy as np

study_hours = [1, 2, 3, 4, 5]
marks = [40, 50, 60, 70, 80]

corr = np.corrcoef(study_hours, marks)
print(corr)
```

```python
import pandas as pd

data = {
    "Hours": [1, 2, 3, 4, 5],
    "Marks": [40, 50, 60, 70, 80]
}

df = pd.DataFrame(data)

print(df.corr())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "Hours": [1,2,3,4,5],
    "Marks": [40,50,60,70,80],
    "Sleep": [8,7,6,6,5]
})

sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
