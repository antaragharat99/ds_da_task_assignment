```python
import matplotlib.pyplot as plt

fruits = ["Apple","Orange","Grapes","Mango","Pineapple"]
sizes = [10000,20000,30000,40000,50000]

plt.pie(sizes,labels=fruits)

plt.show()
```

```python
import matplotlib.pyplot as plt

fruits = ["Apple","Orange","Grapes","Mango","Pineapple"]
sizes = [10000,20000,30000,40000,50000]
colors = ["red","orange","purple","gold","yellow"]
explore = [0.1,0,0,0,0.15]
shadow = True
startangle = 45

plt.pie(sizes,labels=fruits,colors=colors,explode=explore,shadow=shadow,startangle=startangle,autopct="%1.2f%%")


plt.show()
```

```python
import matplotlib.pyplot as plt

fruits = ["Apple","Orange","Grapes","Mango","Pineapple"]
sizes = [10000,20000,30000,40000,50000]
colors = ["red","orange","purple","gold","yellow"]
explore = [0.1,0,0,0,0.15]
shadow = True
startangle = 45

plt.pie(sizes,labels=fruits,colors=colors,explode=explore,shadow=shadow,startangle=startangle,autopct="%1.2f%%",
        radius=1.2,
        labeldistance=1.2)


plt.show()
```

```python
import matplotlib.pyplot as plt

name = ["Antara","Ram","Maruti","Shyam"]
sizes = [200,459,500,100]

plt.title("Bar Plot")
plt.xlabel("Name")
plt.ylabel("Party Cost")
plt.title("Cost Estimate")
plt.bar(name,sizes,color="orange",width=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Antara","Ram","Maruti","Shyam"]
y= [9000,2000,4000,3500]
color = ["orange","yellow","green","blue"]

plt.title("party cost")
plt.xlabel("name")
plt.ylabel("cost")

plt.bar(x,y,color=color)
plt.show()
```
