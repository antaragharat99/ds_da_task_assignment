```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
text2= "Hello ! This is a simple code for genertaing output. NLP is fun and intersting"
print(text2)
```

```python
### Sentence Tokenisation
#From a paragraph  = splitting text into sentences
# Analyse - Sentence Meaning
## Summarisation
## Chatbots
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(text2)
print(sentences)
```

```python
### Word Tokenisation = splitting sentences into words
## Use -
# Counting words
# Text Classification
# Sentiment Analysis
from nltk.tokenize import word_tokenize
words = word_tokenize(text2)
print(words)
```

```python
### Stopwords are words which does not having any meaning
# is, for, this, the, a , an ..
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
filtered_words = []
for w in words:
    if w not in stop_words:
        filtered_words.append(w)
print(filtered_words)
```

```python
fltr_word = [w for w in words if w not in stop_words]  ### Much faster approach
print(fltr_word)
```

```python
## Removing Punctuation
import string  ### Inbuilt library in Python
print(string.punctuation)
```

```python
cleaned_words = []
for p in filtered_words:
  if p not in string.punctuation:
    cleaned_words.append(p)
print(cleaned_words)
```

```python
clnd_words = [p for p in filtered_words if p not in string.punctuation]
print(clnd_words)
```

```python
### Converting text to Lower Case -
# "NLP" and "nlp" should be treated in same way
lowercase_words = [w.lower() for w in cleaned_words]
print(lowercase_words)
```

```python
import pandas as pd
data = {
    "CustomerID":[1,2,3,4],
    "Feedbacks":[
        "The product is amzing !I really loved it.",
        "Customer service was slow, but the staff was polite.",
        "Very bad experience. I will not recommend this.",
        "Good quality and fast delivery !"
    ]
}
# print(data)
df = pd.DataFrame(data)
print(df)
```

```python
def pre_process_text(text):

  ## Word Tokenize
  words = word_tokenize(text)

  ## Remove Stopwords
  words = [w for w in words if w not in stop_words]

  ## Removing punctuation
  words = [w for w in words if w not in string.punctuation]

  ## Converting to lower case
  words = [w.lower() for w in words]

  return words
```

```python
df['Cleaned_feedbacks'] = df['Feedbacks'].apply(pre_process_text)
print(df)
```
