```python
import numpy as np
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[7,8,9],[10,11,12]])
print("Matrix A : \n",A)
print("Matrix B : \n",B)
```

```python
print("Addition: \n", A + B)
```

```python
print("Subtraction: \n", B - A)
```

```python
print("Mutiplication: \n", A * B)
```

```python
print("\nDot Product:\n", np.dot(A, B.T))
```

```python
print("\nTranspose:\n", A.T)
```
