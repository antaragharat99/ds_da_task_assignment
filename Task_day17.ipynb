```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

data = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6],
    "result": [0, 0, 0, 1, 1, 1]
})

X = data[['hours']]
Y = data['result']

model = LogisticRegression().fit(X, Y)

hours = int(input("Enter your study hours: "))

input_df = pd.DataFrame({"hours": [hours]})

print(model.predict(input_df))
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

data = pd.DataFrame({
    "weather": [0, 0, 1, 1, 1],
    "play": ["No", "No", "Yes", "Yes", "Yes"]
})

X = data[['weather']]
Y = data['play']

model = LogisticRegression().fit(X, Y)

condition = int(input("Enter weather condition: "))

predicted_value = model.predict([[condition]])

print("Prediction:", predicted_value[0])
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

data = pd.DataFrame({
    "weather": [2000, 800, 900, 1200, 1400],
    "quality": ["very good", "very bad", "bad", "good", "good"]
})

model = RandomForestClassifier()

X = data[['weather']]
Y = data['quality']

model.fit(X, Y)

condition = int(input("Enter weather data: "))

predicted_value = model.predict([[condition]])

print("Prediction:", predicted_value[0])
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

df = pd.DataFrame({
    'Age': [16, 22, 28, 35, 40],
    'Savings': [100, 400, 300, 200, 800],
    'Buys': ['No', 'Yes', 'Yes', 'No', 'Yes']
})

X = df[['Age', 'Savings']]
y = df['Buys']

model = RandomForestClassifier(n_estimators=10, random_state=42)
model.fit(X, y)

print(model.predict([[30, 500]])[0])
```
