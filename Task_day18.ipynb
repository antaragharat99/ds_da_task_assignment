```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = {
    'Age': [16, 22, 28, 35, 40],
    'Experience': [1, 5, 2, 7, 3],
    'Income': [10000, 40000, 30000, 20000, 80000],
    'Voted': [0, 1, 1, 0, 1],
}
df = pd.DataFrame(data)
# display(df)

X = df[['Age','Experience','Income']]
Y = df['Voted']

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=42)

model = RandomForestClassifier()
model.fit(X_train,y_train)

result = model.predict(X_train)
print(result)
```

```python
X_test
```

```python
X_train
```

```python
acc = accuracy_score(y_test, result)

print(acc * 100)
```

```python
from sklearn.cluster import KMeans
import pandas as pd

data = {
    'Math': [20, 22, 90, 88, 25, 92],
    'Science': [25, 27, 85, 92, 30, 88]
}

df = pd.DataFrame(data)

model = KMeans(n_clusters=2, random_state=42)
df['Group'] = model.fit_predict(df[['Math', 'Science']])

display(df)
# print(Kmeans.cluster_centers_)
# print(Kmeans.labels_)
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

data = np.array([[1,2],[1,4],[1,0],[10,8],[10,9],[10,0]])
linked = linkage(data, method='ward')
plt.figure(figsize=(8,4))
dendrogram(linked,
           labels=['A','B','C','D','E','F'],
           distance_sort='ascending',
           show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distances')
plt.show()
```
