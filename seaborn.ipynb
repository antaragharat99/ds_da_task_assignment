```python
# Basic imports
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 25, 40]

sns.lineplot(x=x, y=y)
plt.title("Line Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

categories = ["Math", "Science", "English"]
marks = [80, 70, 90]

sns.barplot(x=categories, y=marks)
plt.title("Bar Plot Example")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5]
marks = [35, 40, 50, 65, 75]

sns.scatterplot(x=hours, y=marks)
plt.title("Scatter Plot Example")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40, 50]

sns.histplot(data, bins=5)
plt.title("Histogram Example")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [35, 40, 45, 50, 55, 60, 95]

sns.boxplot(x=marks)
plt.title("Box Plot Example")
plt.show()
```
