```python
# 4. Email column data in lowercase
df["Email"] = df["Email"].str.lower()
```

```python
# 5. Make all column names first letter uppercase rest lowercase
df.columns = df.columns.str.title()
```

```python
# 6. Why lower case not change original data?
df["Email"].str.lower()
```

```python
# 7. Need all first name start with D
df[df["First_Name"].str.startswith("D")]
```

```python
# 8. Combine first name & last name and make Full_Name column
df["Full_Name"] = df["First_Name"] + " " + df["Last_Name"]
```

```python
# 9. Delete data of Donald
df = df[df["First_Name"] != "Donald"]
```

```python
# 10. Change position of column
col = df.pop("First_Name")
df.insert(0, "First_Name", col)
```

```python
# 11. Print all people salary greater than 10000
df[df["Salary"] > 10000]

# Count all people salary greater than 10000
df[df["Salary"] > 10000].shape[0]
```

```python
# 12. Need all people hired in 2005
df[df["Hire_Date"].str.contains("05")]
```

```python
# 13. Need all people hired in Aug
df[df["Hire_Date"].str.contains("AUG")]
```
