```python
from sklearn.preprocessing import LabelEncoder

data = ["Yes", "No", "Yes", "No"]

le = LabelEncoder()
encoded = le.fit_transform(data)

print(encoded)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.DataFrame({
    "Gender": ["Male", "Female", "Female", "Male"]
})

le = LabelEncoder()
df["Gender_Encoded"] = le.fit_transform(df["Gender"])

print(df)
```
