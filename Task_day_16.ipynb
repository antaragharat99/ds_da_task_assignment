```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Training Data
data = pd.DataFrame({
    "Hours": [ 4, 5, 6],
    "Score": [60, 70, 80]
})

# Split X (input) and Y (output)
X = data[['Hours']]
Y = data['Score']

# Train the Model
model = LinearRegression().fit(X, Y)

# Take input from user
hours = float(input("Enter your study hours: "))

# Predict score
predicted_score = model.predict(pd.DataFrame({"Hours": [hours]}))

print(f"If you study for {hours} hours, you will score approximately {predicted_score[0]:.2f}%")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({
    "Cores": [10, 20, 30, 90, 100],
    "Score": [0, 0, 1, 1, 1]  # 0 = Flop, 1 = Hit
})

X = data[['Cores']]
Y = data['Score']

model = LinearRegression().fit(X, Y)

cores = float(input("Enter your movie collection (crores): "))

predicted_value = model.predict(pd.DataFrame({"Cores": [cores]}))[0]

prediction = 1 if predicted_value >= 0.5 else 0

if prediction == 1:
    print("Movie is HIT")
else:
    print("Movie is FLOP")
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = pd.DataFrame({
    "weather": [0, 0, 1, 1, 1],
    "play": ["No", "No", "Yes", "Yes", "Yes"]
})

X = data[['weather']]
Y = data['play']

model = DecisionTreeClassifier().fit(X, Y)

condition = int(input("Enter weather condition: "))

predicted_value = model.predict([[condition]])

print("Prediction:", predicted_value[0])
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = pd.DataFrame({
    "size": [1000, 800, 900, 1200, 1400],
    "location": ["high", "low", "low", "very high", "very high"]
})

model = DecisionTreeClassifier()

X = data[['size']]
Y = data['location']

model.fit(X, Y)

condition = int(input("Enter size of your house: "))

predicted_value = model.predict([[condition]])

print("Prediction:", predicted_value[0])
```
