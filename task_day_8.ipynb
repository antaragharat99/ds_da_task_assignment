```python
for num in range(1, 6):
    if num == 3:  # Stop the loop when num is 3
        break
    print(num)
print("Loop stopped")
```

```python
num = 1
while num <= 5:
    if num == 4:  # Stop the loop when num is 4
        break
    print(num)
    num += 1
print("Loop stopped")
```

```python
text = "  Hello Python  "
text1 = "  Easy To Use  "
print(text.strip())  # Output: Hello Python
print(text.upper())  # Output: HELLO PYTHON
print(text.replace("Python", "World"))  # Output: Hello World
print(text.lower())  # Output: HELLO PYTHON
print("-".join([text, text1]))
```
