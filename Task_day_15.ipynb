```python
fruits = ["apple","APPLE","Apple"]
new_fruits = []
for fruit in fruits:
    new_fruits.append(fruit.lower())
print(new_fruits)
```

```python
fruits = ["apple","APPLE","Apple"]
new_fruits = [fruit.lower() for fruit in fruits]
print(new_fruits)
```
